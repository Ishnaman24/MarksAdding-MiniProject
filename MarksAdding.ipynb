```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/project.csv")
df.head()
#creating a data frame
```

```python
df.count
#counting number of rows and columns
```

```python
df['Sum'] = df.groupby(['Roll Num'])['Marks'].transform('sum')
print(df.drop_duplicates(subset=['Roll Num'], keep='last'))
#using the groupby function to comibine similar roll numbers and then adding their corresponding marks 
#by the use of transform('sum')
#also duplicates have been dropped by using the drop_duplicates function
```
